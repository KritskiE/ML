```python
#aes drive link https://drive.google.com/file/d/1-9D8035oQitHRRpnRfPMqPReSYzwJsJ-/view?usp=sharing
#mars drive link https://drive.google.com/file/d/1-BgM5x2pdcWLZcP03WC90AwghiyeDGgf/view?usp=sharing
df_aes=pd.read_csv('/content/drive/MyDrive/aes.csv',header=None)
df_mars=pd.read_csv('/content/drive/MyDrive/mars.csv',header=None)
dfa = df_aes.drop([0],axis=1)
dfm = df_mars.drop([0],axis=1)
```

```python
za= (dfa.to_numpy()).astype(int)
zm = (dfm.to_numpy()).astype(int)
```

```python
y = np.zeros((30,1))
```

```python
def entropija(X,blok):
  y=np.zeros(((len(X)),int((len(X[0,:])/blok))))
  for i in range(len(X)):
    p=0;vel=blok
    for j in range(len(y[0,:])):
      y[i,j]=entropy(X[i,p:vel],base=2)
      p+=blok;vel+=blok
  df=pd.DataFrame(y)
  return df
```

```python
aes_entropija = entropija(za,2048)
mars_entropija = entropija(zm,2048)
```

```python
df= pd.concat([aes_entropija, mars_entropija])
```

```python
Oz = df.to_numpy()
Oz.shape;t=np.concatenate((np.full(30,0),np.full(30,1)),axis=None)
X_train, X_test, y_train, y_test = train_test_split(Oz, t, test_size=0.3, random_state=121, shuffle=True)
```

```python
#RF
from sklearn import metrics
clf=RandomForestClassifier(n_estimators=300)
clf.fit(X_train,y_train)
y_predict=clf.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test,y_predict))
```

```python
#XGboost
xg = XGBClassifier(max_depth=2,n_estimators=20)
xg.fit(X_train,y_train)

y_pred = xg.predict(X_test)
print("Accuracy: ",metrics.accuracy_score(y_test,y_pred) )
```

```python
#QDA
qda = QuadraticDiscriminantAnalysis()
qda.fit(X_train,y_train)
y_pred=qda.predict(X_test)
accuracy=accuracy_score(y_pred, y_test)
print("Accuracy: ", metrics.accuracy_score(y_pred,y_test))
```
